Creating DIctionary using GENSIM

```python
import gensim
from gensim import corpora
from pprint import pprint

# How to create a dictionary from a list of sentences?
documents_1 = ["Natural Language Processing is among the hottest topic in the field of data science",
             "Companies are putting tons of money into research in this field",
             "Everyone is trying to understand Natural Language Processing and its applications to make a career around it",
             "Every business out there wants to integrate it into their business somehow"]

# Tokenize(split) the sentences into words
texts = [[text for text in doc.split()] for doc in documents_1]

# Create dictionary             
dictionary = corpora.Dictionary(texts)

# Get information about the dictionary
print(dictionary)
print(dictionary.token2id)
```

Creating BOW model through GENSIM

```python
from gensim import models
import numpy as np

documents = ["Natural Language Processing is among the hottest topic in the field of data science",
             "Companies are putting tons of money into research in this field",
             "Everyone is trying to understand Natural Language Processing and its applications to make a career around it"]

# Create the Dictionary and Corpus
mydict = corpora.Dictionary([simple_preprocess(line) for line in documents])
corpus = [mydict.doc2bow(simple_preprocess(line)) for line in documents]

# Show the Word Weights in Corpus
for doc in corpus:
    print([[mydict[id], freq] for id, freq in doc])


# Create the TF-IDF model
tfidf = models.TfidfModel(corpus, smartirs='ntc')

# Show the TF-IDF weights
for doc in tfidf[corpus]:
    print([[mydict[id], np.around(freq, decimals=2)] for id, freq in doc])
```
